```python
import gensim
import numpy as np
import pprint as pp
```

```python
import gensim.downloader
list(gensim.downloader.info()['models'].keys())
```

```python
your_model_code = 'glove-twitter-25' # select among the model code aboves
model = gensim.downloader.load(your_model_code) # download and load the model. It can take some time
```

```python
your_word = 'french'
model[your_word]
```

```python
model.most_similar(your_word)
```

```python
import matplotlib.pyplot as plt

plt.plot(model[your_word]) # blue
plt.plot(model['korean']) # green
plt.plot(model['hungry']) # orange
plt.plot(model['walking']) # red
```
